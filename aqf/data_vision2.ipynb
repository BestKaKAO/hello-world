```python
import numpy as np 
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2000)
y=np.random.randn(20,2).cumsum(axis=0)
y
```

```python
plt.figure(figsize=(7,4))
plt.plot(y,lw=1.5)
plt.grid(True)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('2 dimension plot')
#这种情况下 我想添加图例是没办法加的 因为我也不知道哪个是第一列哪个是第二列 所以要用下面这种方法
```

```python
plt.figure(figsize=(7,4))
plt.plot(y[:,0],lw=1.5,label='Line1')
plt.plot(y[:,1],lw=1.5,label='Line2')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('2 dimension plot')
```

```python
y[:,0]=y[:,0]*100

plt.figure(figsize=(8,6))
plt.plot(y[:,0],lw=1.5,label='Line1')
plt.plot(y[:,1],lw=1.5,label='Line2')

plt.grid(True)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('2 dimension plot')
```

```python
fig1,ax=plt.subplots(figsize=(10,6))  #返回一个fig对象，和坐标轴对象

plt.plot(y[:,0],lw=1.5,label='Line1')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=8)

ax2=ax.twinx()  #复制x轴
plt.plot(y[:,1],'r',lw=1.5,label='Line2')
plt.legend()
plt.ylabel('y2')
```

```python
#绘制两个分开的子图
plt.figure(figsize=(10,8))#声明一个画布对象
plt.subplot(211)
plt.plot(y[:,0],lw=1.5,label='1st')
plt.ylabel('y1')
plt.title('Sub Plot')
plt.legend(loc=0)
# 上面部分是绘制第一个子图
# 下面部分绘制第二个子图
plt.subplot(212)
plt.plot(y[:,1],lw=1.5,label='2nd')
plt.xlabel('x')
plt.ylabel('y2')
plt.legend(loc=0)
plt.grid(True)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(y[:,0],lw=1.5,label='1st')
plt.grid(True)
plt.legend()
plt.title('1st Data')
plt.xlabel('Index1')
plt.ylabel('Value')

plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,1],width=0.5,color='c',label='2nd')
plt.xlabel('Index2')
plt.title('2nd Data')
```
